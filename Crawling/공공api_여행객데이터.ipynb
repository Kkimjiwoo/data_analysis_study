```python
import requests
import json
import pandas as pd
```

```python
# 발급 받은 api key 입력하세요.

api_key = ''
```

```python
# 사용할 api url 설정
api_url = 'http://apis.data.go.kr/B551011/DataLabService/metcoRegnVisitrDDList?serviceKey=ServiceKey&numOfRows=10&pageNo=1&MobileOS=ETC&MobileApp=TestApp&_type=json'
```

```python
# 요청 파라미터 설정합니다.
params = {
  'numOfRows' : '10', 
  'pageNo' : '1', 
  'MobileOS' : 'ETC',
  'MobileApp' : 'AppTest',
  'serviceKey' : api_key,
  '_type' : 'json',
  'startYmd':'20210513',
  'endYmd':'20210513'
}
```

```python
# API 요청 보내기
response = requests.get(api_url, params = params)
```

```python
data = response.json()
print(json.dumps(data, indent=4, ensure_ascii = False))
```

```python
#  응답 데이터 확인하기 

if response.status_code == 200: 
  try: 
    data = response.json()
    # json(key, value 형태 파일 >> dictionary 변경)로 parsing 된 데이터 
    print(json.dumps(data, indent=4, ensure_ascii=False))
    # indent=4 : 4칸 들여쓰기 사용 
    # ensure_ascii=False : ASCII 문자 비사용 (한글이 제대로 출력)
    
    # json 데이터에서 필요한 필드만 추출 >> 리스트 저장 
    items = data['response']['body']['items']['item']
    parsed_data = []
    
    for item in items:
      parsed_data.append({
        'areaCode': item['areaCode'],
        'areaNm' : item['areaNm'],
        'daywkDivCd':item['daywkDivCd'],
        'touDivNm': item['touDivNm'],
        'touNum': item['touNum'],
        'baseYmd': item['baseYmd']
      })
      
    # 추출한 데이터를 판다스 데이터프레임으로 변환
    df = pd.DataFrame(parsed_data)
    print(df)
    
  except json.JSONDecodeError:
    print("Error 발생")
```
