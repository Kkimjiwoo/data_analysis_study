```python
import os
import json
import pandas as pd

# JSON 파일들이 들어있는 상위 디렉토리 경로
json_dir = '/content/drive/MyDrive/Final_project_2조/01. 데이터 수집/data/train'

# 데이터를 저장할 빈 리스트 초기화
data = []

# 상위 디렉토리 내의 모든 하위 폴더를 순회
for root, dirs, files in os.walk(json_dir):
    for filename in files:
        if filename.endswith(".json"):  # 파일이 .json 확장자를 가지는지 확인
            file_path = os.path.join(root, filename)

            # JSON 파일을 열고 데이터를 읽기
            with open(file_path, 'r', encoding='utf-8') as f:
                data_json = json.load(f)

            # JSON 데이터에서 annotations 부분을 추출
            annotations = data_json.get("annotations", {})

            # 파일명에서 ID와 location을 추출
            parts = filename.split('_')
            ID = parts[0]
            location = parts[2]

            # 데이터를 리스트로 저장
            data_row = {"ID": ID, "location": location}
            data_row.update(annotations)

            # 리스트에 추가
            data.append(data_row)

# 리스트를 데이터프레임으로 변환
df = pd.DataFrame(data)

# 추출된 모든 annotation 키 값들이 컬럼에 모두 존재하도록 하기 위해
all_columns = ['ID', 'location', 'acne', 'chin_sagging', 'r_cheek_pore', 'r_cheek_pigmentation',
               'forehead_wrinkle', 'r_perocular_wrinkle', 'l_cheek_pore', 'glabellus_wrinkle',
               'lip_dryness', 'forehead_pigmentation', 'l_cheek_pigmentation', 'l_perocular_wrinkle']

# 누락된 컬럼이 있는 경우 NaN 값으로 추가
df = df.reindex(columns=all_columns)

# ID와 location을 기준으로 그룹화한 후, 각 그룹에서 첫 번째 값만 남기기
df_grouped = df.groupby(['ID', 'location'], as_index=False).agg('first')

# ID와 location을 기준으로 정렬
df_grouped = df_grouped.sort_values(by=["ID", "location"]).reset_index(drop=True)

# 결과 확인
print(df_grouped)
```

```python
import os
import json

# JSON 파일들이 들어있는 디렉토리 경로
directory = "/content/drive/MyDrive/Final_project_2조/01. 데이터 수집/data/dataset/train/json/01_40"

# annotation에 있는 모든 키를 저장할 집합(set) 초기화
all_keys = set()

# 디렉토리에 있는 모든 파일들 순회
for filename in os.listdir(directory):
    if filename.endswith(".json"):  # 파일이 .json 확장자를 가지는지 확인
        file_path = os.path.join(directory, filename)

        # JSON 파일을 열고 데이터를 읽기
        with open(file_path, 'r', encoding='utf-8') as f:
            data = json.load(f)

        # JSON 데이터에서 annotations 부분을 추출
        annotations = data.get("annotations", {})

        # annotation의 키 값들만 추출해서 집합에 추가
        all_keys.update(annotations.keys())

# 모든 키를 리스트로 변환 후 출력
all_keys_list = list(all_keys)
print("추출된 모든 키:", all_keys_list)
```
