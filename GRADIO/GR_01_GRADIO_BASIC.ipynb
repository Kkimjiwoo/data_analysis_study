```python
# gradio 통해 이름 입력 >> 인사하는 페이지 구성

import gradio as gr

def greet(name):
  return 'hello, ' + name + '!!^_^'

demo = gr.Interface(fn = greet, inputs = 'text', outputs = 'text')
# fn : 실행할 함수 지정
# inputs: 입력 타입 지정
# outputs: 출력 타입 지
demo.launch(share=True)


# gr.load와 gr.Interface의 차이
# gr.load의 경우 이미 Hugging Face Model Hub에 업로드된 사전 훈련된 모델을 불러와 사용
# gr.Interface의 경우 사용자가 직접 정의한 함수로 사용 / 커스터마이징 가능
```

```python
# 문자열을 뒤집는 웹
import gradio as gr

def flip_text(x):
  return x[::-1]

# gr.Blocks() : 다양한 UI 요소를 구성하고 배치할 수 있는 인터페이스
# gr.Blocks 컨텍스트 매니저는 블록 내의 UI 요소를 관리
with gr.Blocks() as blocks_test:
  # 설명 텍스트 추가
  gr.Markdown("뒤집고 싶은 문자열을 입력하고, 실행을 눌러 결과를 확인하세요.")
  with gr.Row():
    # 텍스트 박스 2개와 버튼을 한 줄에 배치
    inp = gr.Textbox(placeholder = '뒤집고 싶은 문자열')
    out = gr.Textbox()
  btn = gr.Button('실행')
  # 함수를 지정하고, 입력 및 출력 요소를 연
  btn.click(fn = flip_text, inputs =inp, outputs = out)

# 메서드 호출 후 인터페이스 실
blocks_test.launch()


# with 구문
# > 파일 입출력, 네트워크 연결, 데이터베이스 연결 등 효과적으로 관리
```

펠린드롬 검사기 만들기

```python
import gradio as gr

def gr_is_palindrom(text):
  reverse_text = text[::-1]
  # 문자열을 뒤집었을 때와 같은지를 비교
  if text == reverse_text:
    return (reverse_text, '펠린드롬입니다')
  else:
    return (reverse_text, '펠린드롬이 아닙니다')
    # 반환 값이 두개,

# gr_is_palindrom('뉴진스')
interface = gr.Interface(fn = gr_is_palindrom,
                         inputs = 'text',
                         outputs = ['text', 'text'])
                        # 리턴 값이 두개이므로 두개의 outputs
                        # 결과만 보여지고 싶다면 반환 값을 변경해야함
interface.launch(share=True)
```
