```python
# BMI 계산기 1 - input 여러 개 다루기

import gradio as gr

def bmi(name, height, weight):
  bmi_val = round(weight / (height / 100) ** 2,2)
  return (f'안녕하세요.{name}님! 당신의 BMI는 {bmi_val}입니다')

# inputs 타입 형식을 지정해 줘야 함
# input 값에 대한 설명은 함수 입력 변수로 저장됨
gr.Interface(fn=bmi, inputs=['text', 'number', 'number'], outputs='text').launch()
```

```python
# BMI 계산기 2 - input 여러 개 다루기

import gradio as gr

def bmi(name, height, weight):
  bmi_val = round(weight / (height / 100) ** 2,2)
  emotion = ""
  if bmi_val > 23:
    emotion = '살 좀 빼시죠ㅜㅜ'
  else:
    emotion = '살 좀 찌셔도 될 듯😊'
  return (f'안녕하세요.{name}님! 당신의 BMI는 {bmi_val}입니다', emotion)


gr.Interface(fn=bmi,
             inputs=['text', 'number', 'number'],
             outputs=['text', 'text']).launch()
```

```python
import gradio as gr

def bmi(name, height, weight, diet):
    bmi_val = round(weight / (height / 100) ** 2, 2)
    emotion = ""
    if bmi_val > 23:
        emotion = '제발 살 좀 빼시죠ㅜㅜ, 너의 몸이 너무 아프자나......'
    else:
        emotion = '살 좀 찌셔도 될 듯😊'

    if diet:
        diet_message = '잘하고 있군..꾸준히 운동해라.'
    else:
        diet_message = '너의 몸은 지금 점점 풍선처럼 불어가고 있어...너 어디 뭐 살찌러 가니?'

    return (f'안녕하세요.{name}님! 당신의 BMI는 {bmi_val}입니다', emotion, diet_message)


interface = gr.Interface(
    fn=bmi,
    inputs=['text',
            gr.Slider(1, 300, label='키(cm)'),
            gr.Slider(1, 150, label='몸무게(kg)'),
            gr.Checkbox(label='다이어트 여부'),

            ],
    outputs = ['text','text','text'],
    description='문제가 생기면 Flag 를 누르세요.',
    examples=[['지우', 164, 40, True],
        ['병혁', 173.9, 75, False]],
    flagging_options=['good','bad','error'],
    title='BMI 계산기',
    live=True,
    theme=gr.themes.Soft()
)

interface.launch()
```
