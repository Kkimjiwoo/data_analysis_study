```python
import gradio as gr

# with 구문을 사용하여 블록을 생성, 블록 안에 여러 요소 추가/정의
# with 구문은 특정 블록 내에서 리소스를 관리, 블록이 끝나면 자동으로 리소스를 해제 또는 정리
# 블록 컨테이너는 여러개의 UI 요소(텍스트, 이미지, 버튼 등)를 포함할 수 있는 구조

# Gradio 블록을 생성하고 demo 변수에 할당
with gr.Blocks() as demo:
  # gr.Markdown을 이용히여 인터페이스에 텍스트 추가
  # '#'을 이용하여 다른 크기의 제목으로 나타냄
  gr.Markdown("## Gradio Demo")
  gr.Markdown("### Hello")
  gr.Markdown("### World")
  gr.Image("http://onepm.kr/src/assets/images/1o_clock.png", width=100)

  demo.launch()
```

```python
# Blocks 사용하기 - Row()
# 가로로 배치

import gradio as gr
# 블록을 생성하고 그 안에 여러 요소를 추가
# gr.Row()를 사용하여 행을 생성
# 행을 사용하면 요소들이 수평적으로 배치
with gr.Blocks() as demo:
    with gr.Row():
        gr.Markdown("#Gradio Gallery")
        gr.Markdown("Hello")
        gr.Markdown("World")
        gr.Image("http://onepm.kr/src/assets/images/1o_clock.png", width=100)

    demo.launch()
```

```python
# Blocks 사용하기 - Column()

import gradio as gr

with gr.Blocks() as demo:
    # 행을 생성하고, 그 안에 두개의 열 생
    with gr.Row():
        with gr.Column(): # 첫 번째 열
            gr.Markdown("#Gradio Gallery")
        with gr.Column(): # 두 번째 열
            gr.Markdown("Hello")
            gr.Markdown("World")
            gr.Image("http://onepm.kr/src/assets/images/1o_clock.png", width=100)

    demo.launch()
```

```python
# Blocks 사용하기 - Box(), Group() 사용하기

import gradio as gr

with gr.Blocks() as demo:
    with gr.Row():
      # 행 내에 그룹 생성
        with gr.Group():
            gr.Markdown("#Gradio Gallery")
        with gr.Group():
            gr.Markdown("Hello")
            gr.Markdown("World")
        with gr.Column():
            gr.Image("http://onepm.kr/src/assets/images/1o_clock.png", width=100)

    demo.launch()
```

```python
# Blocks Tab 사용하기
import numpy as np
import gradio as gr

# 텍스트 뒤집기
def flip_text(x):
    return x[::-1]

# 이미지 뒤집기
def flip_image(x):
    return np.fliplr(x)

with gr.Blocks() as demo:
    gr.Markdown("Flip text or image files using this demo")
    gr.Markdown(" #Flip text")
    text_input = gr.Textbox()
    text_output = gr.Textbox()
    text_button = gr.Button("Flip")
    gr.Markdown(" #Flip Image")
    image_input = gr.Image()
    image_output = gr.Image()
    image_button = gr.Button("Flip")

    # 클릭하면 열리거나 닫리는 패널을 생성
    with gr.Accordion('Open for more!'):
        gr.Markdown("Look at me...")
        # 버튼을 생성하여 뒤집기 버튼을 생성
        text_button.click(flip_text, inputs=text_input, outputs=text_output)
        image_button.click(flip_image, inputs=image_input, outputs=image_output)

demo.launch(share=True)
```
