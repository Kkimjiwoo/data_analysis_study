```python
# 동일 네일의 리뷰 확인할 결과 대부분의 내용 동일
skin = skin.drop_duplicates(subset=('name'), keep='first')
skin.shape
```

```python
# 'review' 열에서 '], ['를 기준으로 문자열 분리
def split_reviews(review_str):
    # 문자열에서 '], ['를 기준으로 분리한 후, 앞뒤의 대괄호 제거
    if isinstance(review_str, str):
        # '], ['를 기준으로 분리
        parts = review_str.split("'], ['")
        # 대괄호 제거
        parts = [part.strip("[]'") for part in parts]
        return parts
    else:
        return []
# 'review' 열에 함수 적용
skin['review'] = skin['review'].apply(split_reviews)
```

```python
skin[['review']]
```

```python
# 각 리뷰를 별도의 행으로 변환하는 함수 정의
def explode_reviews(df):
    # DataFrame의 각 행에 대해 'review' 열을 열의 요소로 분리하여 새로운 행 생성
    return df.explode('review')

# 변환 적용
expanded_skin_data = explode_reviews(skin)
```

```python
# 제품별 리뷰 개수
skin_count = expanded_skin_data[['name','review']].groupby('name').size().reset_index(name='count')
skin_count[skin_count['count'] < 80]
```

```python
expanded_skin_data = expanded_skin_data.drop_duplicates(subset=('name', 'review'))
expanded_skin_data = expanded_skin_data[['name', 'review']]
```

```python
import pandas as pd
# # 모든 열 너비를 표시
#pd.set_option('display.max_colwidth', None)
# # 옵션을 기본값으로 재설정
#pd.reset_option('display.max_colwidth')
```

```python
# 제거한 후, 동일 제품 내 리뷰 개수
expanded_skin_data_count = expanded_skin_data[['name','review']].groupby('name').size().reset_index(name='count')
expanded_skin_data_count[expanded_skin_data_count['count']<80].name.to_list()
```

```python
# 제거할 'name' 목록 정의
names_to_remove = [
    '[1ST 보습케어] 달바 화이트 트러플 퍼스트 아로마틱 토너 155ml',
    '[단독기획] 아누아 어성초 77 수딩 토너 350ml 리필 기획세트(+350ml 리필팩)',
    '[단독기획] 토니모리 원더 세라마이드 모찌 토너 (본품500ml+100ml)',
    '코스알엑스 하이드리움 워터리 토너 280ml'
]

# 'name' 열의 값이 names_to_remove에 포함되지 않은 행들로 새로운 DataFrame 생성
expanded_skin_data = expanded_skin_data[~expanded_skin_data['name'].isin(names_to_remove)]

# 결과 확인
expanded_skin_data.shape
```

```python
def collapse_review(df): # 행으로 변환한 리뷰를 다시 하나의 리스트 형식으로
  return df.groupby('name', as_index = False).agg({'review': lambda x : list(x)})

skin_data = collapse_review(expanded_skin_data)
skin_data.shape
```

```python
skin_data[:1]
```

```python
# 각 제품의 리뷰 수를 80개로 맞추는 함수 정의
def sample_reviews(reviews, n=80):
    if len(reviews) > n:
        return pd.Series(reviews).sample(n=n, random_state=1).tolist()
    else:
        return reviews
```

```python
skin_data['review'] = skin_data['review'].apply(sample_reviews)
```
